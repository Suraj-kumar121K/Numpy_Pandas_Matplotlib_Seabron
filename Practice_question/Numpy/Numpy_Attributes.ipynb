```python
# An array has 12 elements and shape (3, 4).Is it 2D or 3D?
arr = np.arange(1, 13).reshape(3, 4)
print("array:\n",arr)
print("shape: ",arr.shape)
print("Ndim: ",arr.ndim)
```
