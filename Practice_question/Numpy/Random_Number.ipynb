Generate 8 random integers between 1 and 20 and print only even numbers.

```python
arr4 = np.random.randint(1, 20, size=8)
print("Original array:", arr4)

print()
even = arr4[arr4 % 2 == 0]
print("Even numbers:", even)
```

```python
arr5 = np.random.randint(1, 10, size = 6)
print("Array:", arr5)
count = np.sum(arr5 > 5)
print("Numbers greater than 5:", count)
```
